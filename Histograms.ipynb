# Section: Visualising Data
## Histograms

```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
## create data for histogram

# number of data points
n = 1000

# generate data - log-normal distribution
data = np.exp(np.random.randn(n) / 2)
```

```python
plt.plot(data, 's')
```

```python
# show as a histogram

# number of histogram bins
k = 50

plt.hist(data, bins=k);
```

```python
# another option
y, x = np.histogram(data, bins=k)

# bin centres
xx = (x[1:] + x[:-1]) / 2

plt.plot(xx, y)
plt.show()
```
